# The if statement

- We mostly need to check conditions when we are writing programs to change the behavior of the program.

- if statement is a condtional statement which checks a condition and return True or False

- General form for if statement is as following


    if BOOLEAN EXPRESSION:
        STATEMENT
        STATEMENT
        STATEMENT
        .
        .
        .



Following are very important for an if statement:

- We need colon (:) after the conditional statement

- Next statement after the colon must written on the next line

- Next statement after the colon must be indented

- The statements which are indented executed if the condition is True



# Indentation

- We use indent after an if statement. This means that everything indented after an if statement is only executed if the statement is True.

-  If the statement is False, the program will not excecute and skip all indented code and resume at the first line of unintended code.

- We need to maintain indent in order to indicate the scope of the block.


## Example

````python
#Example of "if" statement
subject = 'Math'

if subject =='Math':  #== checking if the statemnt is true
    print('Yayyy!')
    print('I want to say it 50 times')
    print(50 * subject)

````

## Example

````python
number=90

if number==90: #Don't forget to use colon
    print('The number is equal to 90')

if number<70:
    print('The number is less than 70')

if number<=70:
    print("The number is less than equal to 70")

if number>70:
    print('The number is greater than 70')

if number>=70:
    print("The number is greater than equal to 70")


if number!=70:
    print('The number is not equal to 70')
````

# The if-else statement

- Most of the time when we check a condition if the condition is not True you might want to check another condition.
- To do this we can use if-else statement.

## The syntax for an if else statement looks like this:





    if BOOLEAN EXPRESSION:
        STATEMENTS_1        # executed if condition evaluates to True
    else:
        STATEMENTS_2        # executed if condition evaluates to False


- if clause and else clause must be aligned
- Statements must be consistently indented

````python

x=68

if (x>2):
    print("You are right")
    print("You are doing great")
else:
    print("You are wrong")
    print("Try again")

````

````python
#If the subject is Math it will be executed otherwise the statement which is
# indented after else will be printed


subject="Math"

if subject == 'Math':
    print('Yayyy!')
else:
    print("No I want to take another class")
````

# Chained conditionals: The if-elif-else Statement


When we are programming we might have more than 2 conditions. Therefore, we might need to create multiple conditions. These type of conditionals called chained conditional.



```

    if  condition1< condition2:
        First_STATEMENTS
    elif condition1 > condition2:
        Second_STATEMENTS
    elif condition1 = condition2:
        Third_STATEMENTS
    else:
        Fourth_STATEMENTS
```



# Example

````python
x=8

if x<3:
    print('Less than 3')
elif x<10:
    print ('Less than 10')
else:
    print('None')
````

# Nested conditionals
One conditional can also be nested within another. (It is the same theme of composibility, again!) We could have written the previous example as follows:



```
    if x < y:
        STATEMENTS_A
    else:
        if x > y:
            STATEMENTS_B
        else:
            STATEMENTS_C
```



The outer conditional contains two branches. The second branch contains another if statement, which has two branches of its own. Those two branches could contain conditional statements as well.

Although the indentation of the statements makes the structure apparent, nested conditionals very quickly become difficult to read. In general, it is a good idea to avoid them when you can.

Logical operators often provide a way to simplify nested conditional statements. For example, we can rewrite the following code using a single conditional:

## Example

````python
x = 10
y = 10

if x < y:
    print("x is less than y")
else:
    if x > y:
        print("x is greater than y")
    else:
        print("x and y must be equal")
````

# Comparing Strings

- Strings can be compared using the == and != operators
- String comparisons are case sensitive
- Strings can be compared using >, <, >=, and <=
    - Compared character by character based on the ASCII values for each character
    - If shorter word is substring of longer word, longer word is greater than shorter word



````python
# The > operator compares each character in the strings'Mary',  and 'Mark'
#beginning with the first, or leftmost, characters

name1= 'Mary'
name2= 'Mark'
if name1>name2:
    print('Mary is greater than Mark')
else:
    print('Mary is not greater than Mark')
````
